Kmeans (Sklearn)


```python
from sklearn.cluster import KMeans

adjacency = pd.read_csv('/kaggle/input/da324dataminingproject2/adjacency.csv', index_col=0)
seed = pd.read_excel('/kaggle/input/da324dataminingproject2/seed.xlsx', header=None)
attributes = pd.read_excel('/kaggle/input/da324dataminingproject2/attributes.xlsx')

attributes.reset_index(drop=True, inplace=True)

seed_data = []
for i in range(seed.shape[1]):
    indices = seed[i].values
    seed_data.append(attributes.loc[indices].mean().values)
seed_data = pd.DataFrame(seed_data)

attributes_with_seed = pd.concat([attributes, seed_data])

data_for_clustering = attributes_with_seed.iloc[:10952]

random_indices = np.random.choice(
    data_for_clustering.index, size=10, replace=False)
initial_centers = data_for_clustering.loc[random_indices]

kmeans = KMeans(n_clusters=10, init=initial_centers, n_init=1, random_state=42)

kmeans.fit(data_for_clustering)

cluster_labels = kmeans.labels_

attributes['Cluster'] = np.nan
attributes.loc[:10951, 'Cluster'] = cluster_labels

attributes.to_csv('attributes_with_clusters.csv', index=False)

print("Clustering completed for the first 10952 rows, and results saved to 'attributes_with_clusters.csv'.")
```

Kmeans with PCA (on Attributes)


```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

attributes = pd.read_excel('attributes.xlsx')

imputer = SimpleImputer(strategy='mean')
attributes_imputed = pd.DataFrame(
    imputer.fit_transform(attributes), columns=attributes.columns)

attributes_imputed.reset_index(drop=True, inplace=True)

pca = PCA(n_components=0.95)
attributes_pca = pca.fit_transform(attributes_imputed)

data_for_clustering = attributes_pca[:10952]

random_indices = np.random.choice(
    data_for_clustering.shape[0], size=10, replace=False)
initial_centers = data_for_clustering[random_indices]

kmeans = KMeans(n_clusters=10, init=initial_centers, n_init=1, random_state=42)

kmeans.fit(data_for_clustering)

cluster_labels = kmeans.labels_

attributes_imputed['Cluster'] = np.nan
attributes_imputed.loc[:10951, 'Cluster'] = cluster_labels

attributes_imputed.to_csv('attributes_with_clusters.csv', index=False)

print("Clustering completed for the first 10952 rows, and results saved to 'attributes_with_clusters.csv'.")
```

Kmeans with Random Projection (on Attributes)


```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
from sklearn.random_projection import SparseRandomProjection

attributes = pd.read_excel('attributes.xlsx')

imputer = SimpleImputer(strategy='mean')
attributes_imputed = pd.DataFrame(
    imputer.fit_transform(attributes), columns=attributes.columns)

attributes_imputed.reset_index(drop=True, inplace=True)

random_projection = SparseRandomProjection(
    n_components=10)
attributes_rp = random_projection.fit_transform(attributes_imputed)

data_for_clustering = attributes_rp[:10952]

random_indices = np.random.choice(
    data_for_clustering.shape[0], size=10, replace=False)
initial_centers = data_for_clustering[random_indices]

kmeans = KMeans(n_clusters=10, init=initial_centers, n_init=1, random_state=42)

kmeans.fit(data_for_clustering)

cluster_labels = kmeans.labels_

attributes_imputed['Cluster'] = np.nan
attributes_imputed.loc[:10951, 'Cluster'] = cluster_labels

attributes_imputed.to_csv('attributes_with_clusters.csv', index=False)

print("Clustering completed for the first 10952 rows, and results saved to 'attributes_with_clusters.csv'.")
```

```python
from sklearn.random_projection import GaussianRandomProjection
from sklearn.metrics.pairwise import pairwise_distances

n_components = 1  # Number of dimensions after projection
rp = GaussianRandomProjection(n_components=n_components)
X_projected = rp.fit_transform(attributes)

# Assess pairwise distance preservation
pairwise_dist_original = pairwise_distances(attributes)
pairwise_dist_projected = pairwise_distances(X_projected)

# Calculate explained variance
explained_variance_ratio = 1 - (np.sum(pairwise_dist_projected) / np.sum(pairwise_dist_original))
print("Explained variance ratio:", explained_variance_ratio)
```

MLP


```python
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

attributes_with_clusters = pd.read_csv('attributes_with_clusters.csv')

X = attributes_with_clusters.iloc[:, :-1]
y = attributes_with_clusters['Cluster']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=1000, random_state=42)

nan_indices = y_train.index[y_train.isna()]
X_train = X_train.drop(index=nan_indices)
y_train = y_train.drop(index=nan_indices)

mlp_classifier = MLPClassifier(hidden_layer_sizes=(
    100,), activation='relu', solver='adam', random_state=42)
mlp_classifier.fit(X_train, y_train)

predicted_labels = mlp_classifier.predict(X_test)

submission_df = pd.DataFrame(
    {'ID': range(len(predicted_labels)), 'LABEL': predicted_labels})
submission_df.to_csv('submission.csv', index=False)

print("Prediction completed for the remaining 1000 samples, and results saved to 'submission.csv'.")
```
