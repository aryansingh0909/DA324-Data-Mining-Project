## Clustering


#### Kmeans (Sklearn)


```python
from sklearn.cluster import KMeans

adjacency = pd.read_csv('adjacency.csv', index_col=0)
seed = pd.read_excel('seed.xlsx', header=None)
attributes = pd.read_excel('attributes.xlsx')

attributes.reset_index(drop=True, inplace=True)

seed_data = []
for i in range(seed.shape[1]):
    indices = seed[i].values
    seed_data.append(attributes.loc[indices].mean().values)
seed_data = pd.DataFrame(seed_data)

attributes_with_seed = pd.concat([attributes, seed_data])

data_for_clustering = attributes_with_seed.iloc[:10952]

random_indices = np.random.choice(
    data_for_clustering.index, size=10, replace=False)
initial_centers = data_for_clustering.loc[random_indices]

kmeans = KMeans(n_clusters=10, init=initial_centers, n_init=1, random_state=42)

kmeans.fit(data_for_clustering)

cluster_labels = kmeans.labels_

attributes['Cluster'] = np.nan
attributes.loc[:10951, 'Cluster'] = cluster_labels

attributes.to_csv('attributes_with_clusters.csv', index=False)

print("Clustering completed for the first 10952 rows, and results saved to 'attributes_with_clusters.csv'.")
```

#### Spectral Clustering (Scratch)


```python
from sklearn.neighbors import kneighbors_graph


def spectral_clustering(X, n_clusters, n_neighbors=10, gamma=1.0):
    """
    Perform spectral clustering on the given data.

    Parameters:
    X (ndarray): The input data matrix of shape (n_samples, n_features).
    n_clusters (int): The number of clusters to form.
    n_neighbors (int): The number of nearest neighbors to consider for affinity matrix computation.
    gamma (float): The scaling factor for the Gaussian kernel.

    Returns:
    labels (ndarray): The cluster labels for each data point.
    """
    knn_graph = kneighbors_graph(
        X, n_neighbors=n_neighbors, mode='connectivity', metric='euclidean')
    affinity_matrix = knn_graph.toarray()
    affinity_matrix = np.exp(-gamma * (affinity_matrix ** 2))
    degree_matrix = np.diag(np.sum(affinity_matrix, axis=1))
    laplacian_matrix = degree_matrix - affinity_matrix

    eigenvalues, eigenvectors = np.linalg.eigh(laplacian_matrix)

    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(eigenvectors[:, 1:n_clusters + 1])
    labels = kmeans.labels_

    return labels
```

```python
import pandas as pd

adjacency = pd.read_csv('adjacency.csv')
seed = pd.read_excel('seed.xlsx', header=None)
attributes = pd.read_excel('attributes.xlsx')

X = attributes.values

n_clusters = 10
cluster_labels = spectral_clustering(X[:10952], n_clusters)

print("Cluster labels:", cluster_labels)

print(np.unique(cluster_labels))

print(len(cluster_labels))

attributes['Cluster'] = np.nan
attributes.loc[:10951, 'Cluster'] = cluster_labels

attributes.to_csv('attributes_with_clusters.csv', index=False)

print("Clustering completed for the first 10952 rows using Spectral Clustering, and results saved to 'attributes_with_clusters.csv'.")
```

```python
from sklearn.neighbors import kneighbors_graph


def spectral_decomposition(X, n_neighbors=10, gamma=1.0):
    """
    Perform spectral decomposition on the given data.

    Parameters:
    X (ndarray): The input data matrix of shape (n_samples, n_features).
    n_neighbors (int): The number of nearest neighbors to consider for affinity matrix computation.
    gamma (float): The scaling factor for the Gaussian kernel.

    Returns:
    eigenvalues (ndarray): The eigenvalues of the Laplacian matrix.
    eigenvectors (ndarray): The eigenvectors of the Laplacian matrix.
    """
    knn_graph = kneighbors_graph(
        X, n_neighbors=n_neighbors, mode='connectivity', metric='euclidean')
    affinity_matrix = knn_graph.toarray()
    affinity_matrix = np.exp(-gamma * (affinity_matrix ** 2))
    degree_matrix = np.diag(np.sum(affinity_matrix, axis=1))
    laplacian_matrix = degree_matrix - affinity_matrix

    eigenvalues, eigenvectors = np.linalg.eigh(laplacian_matrix)

    return eigenvalues, eigenvectors
```

```python
adjacency = pd.read_csv('adjacency.csv')
seed = pd.read_excel('seed.xlsx', header=None)
attributes = pd.read_excel('attributes.xlsx')

eigenvalues, eigenvectors = spectral_decomposition(adjacency)
k = 10
sorted_indices = np.argsort(-eigenvalues)
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

selected_eigenvectors = sorted_eigenvectors[:, ::k]
```

## MLP

(For Clustering on 1000 Samples)


```python
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

attributes_with_clusters = pd.read_csv('attributes_with_clusters.csv')

X = attributes_with_clusters.iloc[:, :-1]
y = attributes_with_clusters['Cluster']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=1000, random_state=42)

nan_indices = y_train.index[y_train.isna()]
X_train = X_train.drop(index=nan_indices)
y_train = y_train.drop(index=nan_indices)

mlp_classifier = MLPClassifier(hidden_layer_sizes=(
    100,), activation='relu', solver='adam', random_state=42)
mlp_classifier.fit(X_train, y_train)

predicted_labels = mlp_classifier.predict(X_test)

submission_df = pd.DataFrame(
    {'ID': range(len(predicted_labels)), 'LABEL': predicted_labels})
submission_df.to_csv('submission.csv', index=False)

print("Prediction completed for the remaining 1000 samples, and results saved to 'submission.csv'.")
```
