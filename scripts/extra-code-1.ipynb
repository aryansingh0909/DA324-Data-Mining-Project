```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

adjacency = pd.read_csv('adjacency.csv', index_col=0)
seed = pd.read_excel('seed.xlsx', header=None)
attributes = pd.read_excel('attributes.xlsx')

attributes.reset_index(drop=True, inplace=True)

seed_data = []
for i in range(seed.shape[1]):
    indices = seed[i].values
    seed_data.append(attributes.loc[indices].mean().values)
seed_data = pd.DataFrame(seed_data)

attributes_with_seed = pd.concat([attributes, seed_data])

data_for_clustering = attributes_with_seed.iloc[:10952]

random_indices = np.random.choice(
    data_for_clustering.index, size=10, replace=False)
initial_centers = data_for_clustering.loc[random_indices]

kmeans = KMeans(n_clusters=10, init=initial_centers, n_init=1, random_state=42)

kmeans.fit(data_for_clustering)

cluster_labels = kmeans.labels_

attributes['Cluster'] = np.nan
attributes.loc[:10951, 'Cluster'] = cluster_labels

attributes.to_csv('attributes_with_clusters.csv', index=False)

print("Clustering completed for the first 10952 rows, and results saved to 'attributes_with_clusters.csv'.")
```

```python
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

attributes_with_clusters = pd.read_csv('attributes_with_clusters.csv')

X = attributes_with_clusters.iloc[:, :-1]
y = attributes_with_clusters['Cluster']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=1000, random_state=42)

nan_indices = y_train.index[y_train.isna()]
X_train = X_train.drop(index=nan_indices)
y_train = y_train.drop(index=nan_indices)

mlp_classifier = MLPClassifier(hidden_layer_sizes=(
    100,), activation='relu', solver='adam', random_state=42)
mlp_classifier.fit(X_train, y_train)

predicted_labels = mlp_classifier.predict(X_test)

submission_df = pd.DataFrame(
    {'ID': range(len(predicted_labels)), 'LABEL': predicted_labels})
submission_df.to_csv('submission.csv', index=False)

print("Prediction completed for the remaining 1000 samples, and results saved to 'submission.csv'.")
```
